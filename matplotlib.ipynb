## Data Visualizations with python

```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt

%matplotlib.inline

import random
```

```python
x = [i for i in range(1000)]
y = [i**2 for i in x]

plt.style.use('seaborn')
plt.figure(figsize = (10,5))

plt.plot(x,y)
plt.scatter(x,y, c  = range(len(x)), cmap=plt.cm.Reds)
plt.scatter(0,0, c = 'green')
plt.title('line plot')
plt.xlabel('value')
plt.ylabel('square of value')

plt.tick_params(axis='both', which='major', labelsize=14)

plt.show()
```

```python
# generating Random walk
import random

def random_walk(n  =10001):
    x = [0]
    y = [0]
    for i in range(10001):
        import random
        x_dir = random.choice([-1,1])
        x_move= random.choice([0,1,2,3,4])


        y_dir = random.choice([-1,1])
        y_move= random.choice([0,1,2,3,4])

        if y_move == 0 and x_move == 0:
            continue

        x_movement = x[-1] + x_dir*x_move
        x.append(x_movement)
        
        y_movement = y[-1] + y_dir*y_move
        y.append(y_movement)

    r = range(len(x))    

    return x,y,r

x,y,r = random_walk(15000)
```

```python
# Creating a axis
fig, ax = plt.subplots(figsize = (15,10))

# Plotting scatterplot
ax.scatter(x,y, c = r, cmap = plt.cm.Blues,  edgecolors = None, s = 15)

# Plotting single points 
ax.scatter(x[0], y[0], c='green',edgecolors=None, s =  100)
ax.scatter(x[-1], y[-1], c='red',edgecolors=None, s =  100)

ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
```
